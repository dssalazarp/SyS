2.1 explique las semejanzas y diferencias entre las serie de Fourier (exponencial trigonométrica y compacta ) y la transformada de Fourier, transformada de Fourier en tiempo discreto(DTFT) y la transformada directa de Fourier (DFT). considere las diferentes combinaciones entre: espectro continuo y discreto y  señal tiempo continuo y discreto.

#### **respuesta**
la comparación entre la Serie de Fourier (en sus diferentes formas), la Transformada de Fourier, la Transformada de Fourier en Tiempo Discreto (DTFT) y la Transformada Directa de Fourier (DFT) se basa en las relaciones entre señales en tiempo continuo y discreto, tambien en sus espectros continuo o discreto.

---

### **1. Serie de Fourier**
La Serie de Fourier se aplica a **señales periódicas en tiempo continuo**. Su objetivo es representar una señal periódica como una suma infinita de senos y cosenos ponderadas por coeficientes específicos.

#### **Formas principales**:
- **Trigonométrica**: Representa la señal en términos de senos y cosenos.
  $$
  x(t) = a_0 + \sum_{n=1}^{\infty} \left(a_n \cos(n\omega_0 t) + b_n \sin(n\omega_0 t)\right)
  $$
  Donde $(a_n) y (b_n)$son los coeficientes trigonométricos.

- **Exponencial**: Usa exponentes complejos ($(e^{j\omega}$)) para simplificar la representación.
  $$
  x(t) = \sum_{n=-\infty}^\infty c_n e^{jn\omega_0 t}
  $$
  Aquí $(c_n)$ son los coeficientes complejos.

- **Compacta**: Simplifica la forma trigonométrica combinando términos de amplitud y fase.
  $$
  x(t) = A_0 + \sum_{n=1}^\infty A_n \cos(n\omega_0 t + \phi_n)
  $$

#### **Espectro**:
El espectro de la Serie de Fourier es **discreto**, ya que los coeficientes $(c_n), (a_n), (b_n), o (A_n)$ están definidos solo para múltiplos de la frecuencia fundamental$ (\omega_0)$.

---

### **2. Transformada de Fourier (TF)**
La Transformada de Fourier se aplica a **señales no periódicas en tiempo continuo**. Permite analizar señales arbitrarias en términos de sus componentes frecuenciales.

#### **Definición**:
- Transformada:
  $$
  X(f) = \int_{-\infty}^\infty x(t) e^{-j2\pi f t} dt
  $$
- Inversa:
  $$
  x(t) = \int_{-\infty}^\infty X(f) e^{j2\pi f t} df
  $$

#### **Espectro**:
El espectro de la Transformada de Fourier es **continuo**, ya que evalúa la contribución de cada frecuencia (no solo múltiplos de una frecuencia fundamental).

---

### **3. Transformada de Fourier en Tiempo Discreto (DTFT)**
La DTFT se aplica a **señales no periódicas en tiempo discreto**. Es una extensión de la Transformada de Fourier para señales discretas.

#### **Definición**:
- Transformada:
  $$
  X(e^{j\omega}) = \sum_{n=-\infty}^\infty x[n] e^{-j\omega n}
  $$
- Inversa:
  $$
  x[n] = \frac{1}{2\pi} \int_{-\pi}^\pi X(e^{j\omega}) e^{j\omega n} d\omega
  $$

#### **Espectro**:
El espectro de la DTFT es **continuo** en frecuencia, pero está definido sobre un rango finito $(([-\pi, \pi])$), debido a la periodicidad inherente del espectro de las señales discretas.

---

### **4. Transformada Discreta de Fourier (DFT)**
La DFT es una aproximación computacional de la DTFT que trabaja con **señales discretas y finitas**. Es la base del procesamiento digital de señales.

#### **Definición**:
- Transformada:
  $$
  X[k] = \sum_{n=0}^{N-1} x[n] e^{-j \frac{2\pi}{N} kn}
  $$
- Inversa:
  $$
  x[n] = \frac{1}{N} \sum_{k=0}^{N-1} X[k] e^{j \frac{2\pi}{N} kn}
  $$

#### **Espectro**:
El espectro de la DFT es **discreto y periódico**, y tiene un número finito de puntos $(N)$.

### **Semejanzas**
1. Todas descomponen señales en términos de componentes frecuenciales (senos y cosenos o exponenciales complejos).
2. Permiten representar señales en dominios alternativos (tiempo ↔ frecuencia).
3. La DFT y la DTFT están relacionadas con la Transformada de Fourier, pero adaptadas para señales discretas.
4. La Serie de Fourier es un caso especial de la Transformada de Fourier aplicado a señales periódicas.

---

### **Diferencias**
1. **Dominio de la señal**:
   - Serie de Fourier y Transformada de Fourier: Tiempo continuo.
   - DTFT y DFT: Tiempo discreto.

2. **Espectro**:
   - Serie de Fourier y DFT: Discreto.
   - Transformada de Fourier y DTFT: Continuo.

3. **Periodicidad**:
   - La Serie de Fourier se aplica solo a señales periódicas.
   - La Transformada de Fourier, DTFT y DFT funcionan con señales no periódicas (o finitas, en el caso de la DFT).

4. **Aplicaciones**:
   - La Serie de Fourier se usa en sistemas periódicos como análisis de circuitos.
   - La Transformada de Fourier es ideal para señales continuas y análisis espectral.
   - La DTFT y DFT son esenciales en procesamiento digital de señales (audio, imágenes).
